```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

### GRADED 
Use global and groupby aggregates to answer the following questions

    - What is the average tip amount 
    - What is the average bill amount 
    - What is the average tip by gender 
    - What is the average tip on a sunday by gender 
    - Which party size gives the maximum tip as a percent of total bill amount 

```python
# What is the average tip amount?
tip_mean = tips['tip'].mean()
print('1. The average tip amount is:', tip_mean,'.')
```

```python
# What is the average bill amount?
bill_mean = tips['total_bill'] . mean()
print('2. The average bill amount is' , bill_mean, '.')
```

```python
# What is the average tip by gender?
gender = tips. groupby('sex').mean()
male_mean = gender. iloc[0,1]
female_mean = gender. iloc[1,1]
print('3. The average tip by male is:', male_mean, 
      '. The Average tip by female is:', female_mean, '.')
```

```python
# What is the average tip on a sunday by gender?
day = tips. groupby(['day','sex']). mean()   
day.drop(['total_bill','size'], axis = 1, inplace = True)   
day_mean = day.drop(day. index[:6])   
print(day_mean)
print('4. On sunday, average tip by male is 3.220345, average tip by female is 3.367222.')
```

```python
#Which party size gives the maximum tip as a percent of total bill amount
tips['tip_percentage'] = tips['tip'] / tips['total_bill']
tips.groupby(['size'])['tip_percentage'].describe()
```

```python
# Which party size gives the maximum tip as a percent of total bill amount 
print('5. The party size that gives the maximum tip as percent of total bill amount is 2.')
```

```python
# Generate a box plot that shows average bill amount by day of week
# Generate a box plot that shows average bill amount by gender
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
sns.boxplot(x='day',y='total_bill', data=tips)
plt.subplot(1, 2, 2)
sns.boxplot(x='sex',y='total_bill', data=tips)
```

```python
# Using matplotlib
# Generate a box plot that shows average bill amount by day of week
# Generate a box plot that shows average bill amount by gender
%matplotlib inline
import matplotlib.pyplot as plt
Thur = tips.loc[tips['day']=='Thur']
Fri = tips.loc[tips['day']=='Fri']
Sat = tips.loc[tips['day']=='Sat']
Sun = tips.loc[tips['day']=='Sun']
Male = tips.loc[tips['sex']=='Male']
Female = tips.loc[tips['sex']=='Female']

plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
plt.boxplot([Thur['total_bill'],Fri['total_bill'],
             Sat['total_bill'],Sun['total_bill']])
plt.gca().set_xticklabels(['Thur','Fri','Sat','Sun'])

plt.subplot(1,2,2)
plt.boxplot([Male['total_bill'],Female['total_bill']])
plt.gca().set_xticklabels(['Male','Female'])
```

```python
# Using bokeh
# Generate a box plot that shows average bill amount by day of week
# Generate a box plot that shows average bill amount by gender

import holoviews as hv
hv.extension('bokeh')
import seaborn as sns
tips = sns.load_dataset('tips')

title1 = 'Average Bill Amount by Day of Week'
boxwhisker1 = hv.BoxWhisker(tips, ['day'], 'total_bill', label = title1)
boxwhisker1.options(show_legend=False, width=200)

title2 = 'Average Bill Amount by Gender'
boxwhisker2 = hv.BoxWhisker(tips, ['sex'], 'total_bill', label = title2)
boxwhisker2.options(show_legend=False, width=200)

boxwhisker1 + boxwhisker2   
```

### GRADED
    - Using seaborn lmplot(), examine the linear relationship between tips and total_bill across different dimensions to answer the following
    - Do smoker/non-smoker give a higher tip as a percentage of total_bill (which line has a steeper slope)
    - Are tips larger during lunch or dinner (as a percentage of total_bill)
    - Does the gender that tips the most (as a percentage of total_bill) change between lunch and dinner?
        - Answer the question by visual examination of lmplot output

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
```

```python
plt = sns.lmplot(x = 'total_bill', y='tip', data = tips)
plt.fig.suptitle('Bill Vs Tip')
```

```python
#Do smoker/non-smoker give a higher tip as a percentage of total_bill (which line has a steeper slope)
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, markers=['o', 'x'])
print('Non-smokers tip a higher percentage of their bills')
```

```python
#Are tips larger during lunch or dinner (as a percentage of total_bill)
tips['tip/bill'] = tips['tip']/tips['total_bill']
sns.lmplot(x='total_bill', y='tip', hue='time', data=tips, markers=['o', 'x'])
print('Tips are higher for dinner than for lunch.')
```

```python
#Does the gender that tips the most (as a percentage of total_bill) change between lunch and dinner?
sns.lmplot(x='total_bill', y='tip', hue='sex', col='time', data=tips)
print('The highest tips for lunch and dinner came from men, but slightly higher.')
```
