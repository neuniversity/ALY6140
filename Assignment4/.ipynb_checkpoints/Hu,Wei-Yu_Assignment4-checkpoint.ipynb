```python
import seaborn as sns
tips =sns.load_dataset('tips')
tips.head()
```

### GRADED 
Use global and groupby aggregates to answer the following questions

    - What is the average tip amount 
    - What is the average bill amount 
    - What is the average tip by gender 
    - What is the average tip on a sunday by gender 
    - Which party size gives the maximum tip as a percent of total bill amount 
 

```python
# the average of tip amount
tips['tip'].mean()
```

```python
# the average of bill amount
tips['total_bill'].mean()
```

```python
#the average tip by gender 
 tips.groupby('sex')['tip'].mean()
```

```python
# the average tip on a sunday by gender 
tips[tips['day']=='Sun'].groupby('sex')['tip'].mean()
```

```python
 # Which party size gives the maximum tip as a percent of total bill amount 
tp= tips['tip_percent'] = (tips['tip'] / tips['total_bill']) * 100
tp=tips['tp']
tips.iloc[tips['tp'].idxmax()]
```

### GRADED
- Generate a scatter plot of tip amount (y-axis) and total_bill (x-axis)
- Generate the plots using matplotlib, bokeh and seaborn

```python
# generate a scatter plot of tip amount (y-axis) and total_bill (x-axis) 
# with seaborn
sns.scatterplot(x='total_bill', y='tip',data=tips)
```

### GRADED
    - Generate a box plot that shows average bill amount by day of week
    - Generate a box plot that shows average bill amount by gender
    - Show the above two plots side by side
    - Generate the same plots using matplotlib and bokeh

```python
# the plot that shows average bill amount by day of week
#with seaborn
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#a box plot that shows average bill amount by gender
sns.boxplot(x='sex',y='total_bill',data=tips)
```

```python
# show the above two plots side by side
fig, ax =plt.subplots(1, 2)
sns.boxplot(x='day',y='total_bill',data=tips, ax=ax[0])
sns.boxplot(x='sex',y='total_bill',data=tips, ax=ax[1])
fig.show()
```

```python
#matplotlib
#The plot that shows average bill amount by day of week 
plt.figure(figsize=(10,5))

Thur=tips.loc[tips['day']=='Thur',:]
Fri=tips.loc[tips['day']=='Fri',:]
Sat=tips.loc[tips['day']=='Sat',:]
Sun=tips.loc[tips['day']=='Sun',:]
plt.boxplot([Thur['total_bill'],Fri['total_bill'],Sat['total_bill'],Sun['total_bill']])

plt.title('Average bill amount by day of week')
plt.gca().set_xticklabels(['Thur', 'Fri','Sat','Sun'])


```

```python
#The plot that shows average bill amount by gender
Male=tips.loc[tips['sex']=='Male',:]
Female=tips.loc[tips['sex']=='Female',:]
plt.boxplot([Male['total_bill'],Female['total_bill']])

plt.title('Average bill amount by gender')
plt.gca().set_xticklabels(['Male', 'Female'])
```

### GRADED
    - Using seaborn lmplot(), examine the linear relationship between tips and total_bill across different dimensions to answer the following
    - Do smoker/non-smoker give a higher tip as a percentage of total_bill (which line has a steeper slope)
    - Are tips larger during lunch or dinner (as a percentage of total_bill)
    - Does the gender that tips the most (as a percentage of total_bill) change between lunch and dinner?
        - Answer the question by visual examination of lmplot output

```python
#Do smoker/non-smoker give a higher tip as a percentage of total_bill
sns.set(color_codes=True)
sns.lmplot(x='total_bill', y='tip_percent', hue='smoker', data=tips)
```

According to the figure above, it is to say that the percentage of non-smoker is slightly lower than the percentage of smoker. 

```python
#Are tips larger during lunch or dinner ?
sns.lmplot(x='total_bill', y='tip', col='time',hue='time', data=tips)
```

Accroding to the plot, the tips is larger during dinner than during lunch.

```python
#Does the gender that tips the most change between lunch and dinner?
sns.lmplot(x='total_bill', y='tip', col='time',hue='sex', data=tips)
```

According to the figures above, there is no big differenc between the average line of 'Male' and 'Female' during lunch. During dinner, there are a few excpetion spots that are 'Male' which are a lot higher than the average, yet, the average line of 'Female' is slightly higher than the average line of 'Male'at the begining and it went slightly lower than the 'Male' average line when the total bill is higer.It is to say that, 'Male' tends to tip more when the total of the bill is higher than 30.
