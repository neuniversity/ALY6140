```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

### GRADED 
Use global and groupby aggregates to answer the following questions

    - What is the average tip amount 
    - What is the average bill amount 
    - What is the average tip by gender 
    - What is the average tip on a sunday by gender 
    - Which party size gives the maximum tip as a percent of total bill amount 

```python
#Import necessary modules 
import pandas as pd
import numpy as np


#What is the average tip amount 

average_tip_amount  = 5
def func():
    global average_tip_amount
    average_tip_amount = np.mean(tips['tip'])
    print(average_tip_amount) 

func()
```

```python
#What is the average bill amount 

average_bill_amount  = 5
def func2():
    global average_bill_amount
    average_bill_amount = np.mean(tips['total_bill'])
    print(average_bill_amount) 

func2()
```

```python
#What is the average tip by gender 

tips_gender  = 5
def func3():
    global tips_gender
    tips_gender = tips['tip'].groupby(tips['sex']).mean()
    print(tips_gender) 

func3()
```

```python
#What is the average tip on a sunday by gender 

tips_sun = tips.loc[tips['day'] == 'Sun']

tips_gender_sun  = 5
def func4():
    global tips_gender_sun
    tips_gender_sun=tips_sun['tip'].groupby(tips_sun['sex']).mean()
    print(tips_gender_sun) 

func4()
```

### GRADED
    - Generate a scatter plot of tip amount (y-axis) and total_bill (x-axis)
    - Generate the plots using matplotlib, bokeh and seaborn


```python
#Import necessary modules 
import matplotlib.pyplot as plt
%matplotlib inline

from bokeh.plotting import figure, show, output_file
#Output to notebook 
output_file("boxplot.html", title="Boxplot") 

import seaborn as sns


#Set values of 'x', 'y'.
x=tips['total_bill']
y=tips['tip']
```

```python
#Basic scatter plot created by matplotlib

plt.scatter(x, y,alpha=0.3)
plt.xlabel('total_bill')
plt.ylabel('tip amount')
```

```python
#Basic scatter plot created by bokeh

p = figure(plot_width=600, plot_height=400)
p.square(x,y,size=15, color="navy", alpha=0.5,legend = 'scatter-bokeh')
p.legend.location = "bottom_right"
show(p)
```

```python
#Basic scatter plot created by seaborn

p=sns.scatterplot( y="tip", x="total_bill",data=tips)
```

### GRADED
    - Generate a box plot that shows average bill amount by day of week
    - Generate a box plot that shows average bill amount by gender
    - Show the above two plots side by side
    - Generate the same plots using matplotlib and bokeh

```python
#Generate a box plot that shows average bill amount by day of week

sns.boxplot(y="total_bill", x="day", data=tips)
plt.show()
```

```python
#Generate a box plot that shows average bill amount by gender

sns.boxplot(y="total_bill", x="sex", data=tips)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(20,10))
 
plt.subplot(1,2,1)
sns.boxplot(y="total_bill", x="day", data=tips)
 
plt.subplot(1,2,2)
sns.boxplot(y="total_bill", x="sex", data=tips)
 
plt.show()
```

```python
#Show the boxplot show the relationship between differnet gender and days.

sns.boxplot(y="total_bill", x="day",hue="sex", data=tips)
plt.show()
```

### GRADED
    - Using seaborn lmplot(), examine the linear relationship between tips and total_bill across different dimensions to answer the following
    - Do smoker/non-smoker give a higher tip as a percentage of total_bill (which line has a steeper slope)
    - Are tips larger during lunch or dinner (as a percentage of total_bill)
    - Does the gender that tips the most (as a percentage of total_bill) change between lunch and dinner?
    - Answer the question by visual examination of lmplot output

```python
#Show the tips (as a percentage of total_bill) by if the costumers are smokers or not. 
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=tips,palette="Set2")
```

```python
#Show the tips (as a percentage of total_bill) by time in a day. 

sns.lmplot(x="total_bill",y="tip",hue="time",data=tips,palette="Set1")
```

```python
#Show the tips (as a percentage of total_bill) of gender in different times of a day that has the higher tips. 

sns.lmplot(x="total_bill",y="tip",hue="sex",data=tips,palette="Set1")
male_tips=tips.loc[tips['sex'] == 'Male']
sns.lmplot(x="total_bill",y="tip",hue="time",data=male_tips,palette="Set3")
```
