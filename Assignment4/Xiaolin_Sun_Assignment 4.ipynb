### GRADED 
Use global and groupby aggregates to answer the following questions

    - What is the average tip amount 
    - What is the average bill amount 
    - What is the average tip by gender 
    - What is the average tip on a sunday by gender 
    - Which party size gives the maximum tip as a percent of total bill amount 

```python
# - What is the average tip amount
tip_mean = tips.tip.mean()
print('The average tip amount is', tip_mean)

# - What is the average bill amount 
bill_mean = tips.total_bill.mean()
print('The average bill amount is', bill_mean)

# - What is the average tip by gender 
print('======'*10)
print('The average tip by gender is shown as')
tip_bysex = tips.groupby('sex').tip.mean()
print(tip_bysex)

# - What is the average tip on a sunday by gender
print('======'*10)
print('The average tip on a sunday by gender is shown as')
tip_sun = tips[tips['day']=='Sun'].groupby('sex').tip.mean()
print(tip_sun)
# - Which party size gives the maximum tip as a percent of total bill amount
print('======'*10)
tips['tip_percent'] = (tips.tip/tips.total_bill)*100
print('The maximum tip as a percent of total bill amount has a party size of')
tips[tips['tip_percent']==tips['tip_percent'].max()].loc[:,'size']
```

### GRADED
    - Using seaborn lmplot(), examine the linear relationship between tips and total_bill across different dimensions to answer the following
    - Do smoker/non-smoker give a higher tip as a percentage of total_bill (which line has a steeper slope)
    - Are tips larger during lunch or dinner (as a percentage of total_bill)
    - Does the gender that tips the most (as a percentage of total_bill) change between lunch and dinner?
        - Answer the question by visual examination of lmplot output

```python
# Determine the coefficient of the regression.
import numpy as np 
import matplotlib.pyplot as plt 
x = tips.total_bill
y = tips.tip
  
def estimate_coef(x, y): 
    # number of observations/points 
    n = np.size(x) 
  
    # mean of x and y vector 
    m_x, m_y = np.mean(x), np.mean(y) 
  
    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(y*x) - n*m_y*m_x 
    SS_xx = np.sum(x*x) - n*m_x*m_x 
  
    # calculating regression coefficients 
    b_1 = SS_xy / SS_xx 
    b_0 = m_y - b_1*m_x 
  
    return(b_0, b_1) 

def plot_regression_line(x, y, b): 
    # plotting the actual points as scatter plot 
    plt.scatter(x, y, color = "m", 
               marker = "o", s = 30) 
  
    # predicted response vector 
    y_pred = b[0] + b[1]*x 
  
    # plotting the regression line 
    plt.plot(x, y_pred, color = "g") 
  
    # putting labels 
    plt.xlabel('x') 
    plt.ylabel('y') 
  
    # function to show plot 
    plt.show() 
    
def main(): 
    # observations 
    x = tips.total_bill
    y = tips.tip
  
    # estimating coefficients 
    b = estimate_coef(x, y) 
    print("Estimated coefficients:\nb_0 = {}  \nb_1 = {}".format(b[0], b[1])) 
  
    # plotting regression line 
    plot_regression_line(x, y, b) 
  
if __name__ == "__main__": 
    main() 
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips,height=5, aspect=1.5,)
ax = plt.gca()
ax.set_title("Coefficient = 0.105")
```

# - Do smoker/non-smoker give a higher tip as a percentage of total_bill 
# which line has a steeper slope

```python
plot1 = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
ax = plt.gca()
ax.set_title("Tip vs Total_bill")
plot2 = sns.lmplot(x="total_bill", y="tip_percent", hue="smoker", data=tips, x_jitter=.05)
ax = plt.gca()
ax.set_title("Tip_percent vs Total_bill")
```

The line of smoker has a sleeper slope.
When total bill is below around 23: Smorker give a higher tip as a percent of total bill.
When total bill is higher than 23: Non-smoker give a higher tip as a percent of total bill.

# - Are tips larger during lunch or dinner (as a percentage of total_bill)

It is difficult to make a conclusion about whether tips are larger during lunch or dinner time. It denpend on what "larger" means in the question.

Compare with tips given during lunch time, there are more customers giving tips below 10% of the total bill during dinner time, and there are also more customers giving tips more than 20% of the total bill during dinner time. 

However, we can conclude that there are a few larger tips are given during dinner time.
