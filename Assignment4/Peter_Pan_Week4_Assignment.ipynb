# Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = tips['total_bill']
y = tips['tip']
plt.scatter(x, y,s = x*1.5)
plt.xlabel('total_bill')
plt.ylabel('tip')
```

# Bokeh

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
x = tips['total_bill']
y = tips['tip']
z = x*0.5

p = figure(x_axis_label='total_bill', y_axis_label='tip')
p.circle(
    x, y, size=z,
    color='red')
output_notebook()
show(p)
```

# Seaborn

```python
import pandas as pd
df = pd.DataFrame(tips, columns=["total_bill", "tip"])


plt.figure(figsize=(5,5))
plt.subplot(1, 1, 1)
plt.scatter(df["total_bill"], df["tip"], s = df['total_bill']*1.5)
plt.xlabel('total_bill')
plt.ylabel('tip')
```

# Boxplot average bill by day(seaborn)

```python
sns.boxplot(x="day", y='total_bill', data = tips)
plt.title('average bill amount by day of week')
```

# Boxplot average bill gender(seaborn)

```python
sns.boxplot(x="sex", y='total_bill', data = tips)
plt.title('average bill amount by gender')
```

# Show the above two plots side by side(seaborn)

```python
plt.subplot(1,2,1)
sns.boxplot(x="day", y='total_bill', data = tips)
plt.title('average bill amount by day of week')
plt.subplot(1,2,2)
sns.boxplot(x="sex", y='total_bill', data = tips)
plt.title('average bill amount by gender')
plt.tight_layout()
```

# Boxplot average bill by day(matplotlib)

```python
plt.figure(figsize=(5,5))
Thur = tips.loc[tips['day']=='Thur',:]
Fri = tips.loc[tips['day']=='Fri',:]
Sat = tips.loc[tips['day']=='Sat',:]
Sun = tips.loc[tips['day']=='Sun',:]

plt.boxplot([Thur['total_bill'], Fri['total_bill'], Sat['total_bill'], Sun['total_bill']])
plt.gca().set_xticklabels(['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('average bill amount by day of week')
```

# Boxplot average bill by gender(matplotlib)

```python
# - Generate a box plot that shows average bill amount by gender
# - Show the above two plots side by side
# - Generate the same plots using matplotlib and bokeh
plt.figure(figsize=(5,5))
Male = tips.loc[tips['sex']=='Male',:]
Female = tips.loc[tips['sex']=='Female',:]

plt.boxplot([Male['total_bill'], Female['total_bill']])
plt.gca().set_xticklabels(['Male', 'Female'])
plt.title('average bill amount by gender')
```

# Show the above two plots side by side(matplotlib)

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.boxplot([Thur['total_bill'], Fri['total_bill'], Sat['total_bill'], Sun['total_bill']])
plt.gca().set_xticklabels(['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('average bill amount by day of week')

plt.subplot(1,2,2)
plt.boxplot([Male['total_bill'], Female['total_bill']])
plt.gca().set_xticklabels(['Male', 'Female'])
plt.title('average bill amount by gender')

plt.tight_layout()
```
