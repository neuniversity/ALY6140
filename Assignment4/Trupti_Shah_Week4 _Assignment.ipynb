```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

tips = sns.load_dataset('tips')
tips.head()
```

### GRADED 
Use global and groupby aggregates to answer the following questions

    - What is the average tip amount 
    - What is the average bill amount 
    - What is the average tip by gender 
    - What is the average tip on a sunday by gender 
    - Which party size gives the maximum tip as a percent of total bill amount 

```python
# What is the average tip amount

avg_tip = tips['tip'].mean()
print('The average tip amount - $', round(avg_tip,2))
```

```python
# What is the average bill amount

avg_bill = tips['total_bill'].mean()
print('The average bill amount - $', round(avg_bill,2))
```

```python
# What is the average tip by gender

avgtip_gender = tips.groupby(['sex'])['tip'].mean()
#print(avgtip_gender)
print('The average tip by gender -')
print("Male: {0:.2f}" "\n" "Female: {1:.2f}".format(avgtip_gender[0], avgtip_gender[1]))
```

```python
# What is the average tip on a sunday by gender

avgtip_sunday_gender = tips[tips['day'] == 'Sun'].groupby(['sex'])['tip'].mean()
print('The average tip on a sunday by gender -')
print("Male: {0:.2f}" "\n" "Female: {1:.2f}".format(avgtip_sunday_gender[0], avgtip_sunday_gender[1]))
```

```python
# Which party size gives the maximum tip as a percent of total bill amount 

tip_percent = (tips['tip']/tips['total_bill']) * 100
tips['tip_percent'] = tip_percent
max_tip = tips.iloc[tips['tip_percent'].idxmax()]
print('Party size that gives the maximum tip as a percent of total bill amount -', max_tip['size'])
```

**Using Seaborn -** Generate a box plot that shows average bill amount by day of week

```python
sns.set(style="ticks", palette="pastel")

# Draw a nested boxplot to show bills by day and time
plot = sns.boxplot(x="day", y="total_bill", data=tips)
sns.despine(offset=10, trim=True)

# Calculating the average bill amount by day of week
day_mean = tips.groupby(['day'])['total_bill'].mean().values
mean_labels = [str(np.round(s, 2)) for s in day_mean]

# Printing the average bill amount on the boxplot
pos = range(len(day_mean))
for tick,label in zip(pos, plot.get_xticklabels()):
    plot.text(pos[tick], day_mean[tick] + 0.5, mean_labels[tick], 
            horizontalalignment='center', size='x-small', color='w', weight='semibold')
```

**Using Seaborn -** Generate a box plot that shows average bill amount by gender

```python
# Draw a nested boxplot to show bills by day and time
plot = sns.boxplot(x="sex", y="total_bill",data=tips)

sns.despine(offset=10, trim=True)

day_mean = tips.groupby(['sex'])['total_bill'].mean().values
mean_labels = [str(np.round(s, 2)) for s in day_mean]

pos = range(len(day_mean))
for tick,label in zip(pos,plot.get_xticklabels()):
    plot.text(pos[tick], day_mean[tick] + 0.5, mean_labels[tick], 
            horizontalalignment='center', size='x-small', color='w', weight='semibold')
```

**Using matplotlib-** Generate a box plot that shows average bill amount by day of week

    - Using seaborn lmplot(), examine the linear relationship between tips and total_bill across different dimensions to answer the following
    - Do smoker/non-smoker give a higher tip as a percentage of total_bill (which line has a steeper slope)
    - Are tips larger during lunch or dinner (as a percentage of total_bill)
    - Does the gender that tips the most (as a percentage of total_bill) change between lunch and dinner?
        - Answer the question by visual examination of lmplot output

**Do smoker/non-smoker give a higher tip as a percentage of total_bill (which line has a steeper slope)?**

```python
sns.set(color_codes=True)

# Creating a lmplot
sns.lmplot(x="total_bill", y="tip_percent", hue="smoker", data=tips)
```

**Analysis-** As per the above plots, we observe that smoker have a steeper slope as compared to the non-smoker for total bill vs the tip as a percentage of total_bill. But, we observe that non-smokers pay more tip as a percentage of total_bill when the total bill is above around 22 dollers as compared to smokers. Below the total bill of 22 dollers, smokers are observed to pay more amount of tip.

**Are tips larger during lunch or dinner (as a percentage of total_bill)?**

```python
sns.lmplot(x="total_bill", y="tip", col="time",hue="time", data=tips)
```

```python
# tip as a percentage of total bill

sns.lmplot(x="total_bill", y="tip_percent", hue="time", data=tips)
```

**Analysis:** By observing the above plots, we observe that the number of tips are greater during dinner as compared to lunch. By comparing the slope of the lines in the plot, we can say that the tip as a percentage is greater during lunch as compared to dinner for higher total bill amounts.

**Does the gender that tips the most (as a percentage of total_bill) change between lunch and dinner?**

```python
sns.lmplot(x="total_bill", y="tip_percent", col="time", hue="sex", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip_percent", col="sex", hue="time", data=tips)
```

**Analysis-** From the above plots, we observe a significant change in the tips as a percentage of total_bill given by females between lunch and dinner. Also, males are observed to more number of tips during dinner as compared to females.
