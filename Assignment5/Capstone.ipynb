```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
sns.set_style('whitegrid')
cap=pd.read_csv('cwurData.csv')
df=cap.loc[(cap['year']==2012)]
df
```

```python
df.describe()
```

In 2012, which countries own the best quality of education?

```python
cap1=df.loc[(df['quality_of_education']==1)]
cap1
```

Which university has the best influence?

```python
cap2=df.iloc[:,[1,8]]
cap3=cap2.loc[(cap2['influence']==1)]
cap3
```

As you can see, although the Harvard University has the best influence in the world in 2012, the quality of its education is not the best, which means that the influence will not be decided only by the quality of education.

In 2012, for top 30 universities, describe the relationship between the quality of education and the quality of faculty

```python
import matplotlib.pyplot as plt
df1=df.head(30)
x=df1['quality_of_faculty']
y=df1['quality_of_education']
plt.figure('Drew')
plt.scatter(x,y)
plt.show()
```

According to this scatter chart, the trend is that the quality of faculty will affect the quality of education and the relationship is positive. However, the trend is not very obvious, which means that the quality of education will not decided only by the quality of faculty.

In 2012, for the top 30 universities, describe the relationship between the quality of education and the patents.

```python
cap6=df1.sort_values(by=['patents'])
cap7=cap6.iloc[:,[4,11]]
sns.pairplot(cap7)
```

According to the above charts, the patents will bring some positive influence to the quality of education. Plus, with the increase of the quality of education, the patents will be affected and increased.

In 2012, for the top 30 universities, whether the university which has higher rank will have more influence?

```python
capx=cap.loc[(cap['year']==2012)]
caps=capx.head(30)
sns.swarmplot(x='world_rank',y='influence',data=caps)
```

According to the above chart, we can conclude that the influence will be affected by the rank. The trend is that a higher rank will bring more influence.

In 2012,for top 30 universities, whether the higher requirement of score means the higher quality of education?

```python
cap11=df.iloc[:,[4,12]]
cap12=cap11.head(30)
cap14=cap12.sort_values(by=['quality_of_education'])
sns.barplot(x='quality_of_education',y='score',data=cap14)
```

```python
plt.plot(
    cap14['quality_of_education'],
    cap14['score'],
    'g',
    linewidth=3.0
)
plt.ylabel('score')
plt.xlabel('quality of education')
plt.title('The relationship between score and quality of education')
```

According to the chart, the general trend is that the higher quality of education will require higher score.

In 2012, for top 30 universities, describe the relationship between world rank and the quality of education.

```python
cap13=df1.iloc[:,[0,4]]
sns.jointplot(x='world_rank',y='quality_of_education',data=cap13)
```

According to this chart, we can conclude that an university which has a higher world_rank will own a better quality of education.
