# Reading an Image file
- img.imread() is used to read the image
- After reading, using plt.imshow() the image is displayed
- The image is displayed with values on x and y axis
- To remove the axis for image, we use plt.axis('off')
- Using plt.title() title is inserted for the image

```python
# Displying image 
import matplotlib.pyplot as plt
import matplotlib.image as img

original_image = img.imread("Example_Image.jpg")
plt.title("Original image")
plt.imshow(original_image)
plt.axis('off')
plt.show()

```

# Reading and plotting Binary data
- The Binary file is read using pyabf.ABF() of pyabf
- ABF objects provide access to binary data(ABF data) using sweep number. sweepX gives the data
- Using sweepX, we can access sweep times
- Using sweepY, sweep ADC data is accessed

```python
# Reading Biinary file
import pyabf
import matplotlib.pyplot as plt
    
binary = pyabf.ABF("Example_Binary.abf")
print(binary)
plt.plot(binary.sweepX, binary.sweepY)
#binary.setSweep(3)
#print(binary.sweepY)
plt.ylabel(binary.sweepLabelY)
plt.xlabel(binary.sweepLabelX)
plt.show()
```

# Finding the Relationship between data columns
- Excel file is read using Pandas
- Plot between Country column and Gross Sales Colummn using plt.bar() of matplotlib.pyplot
- plt.show() displays the image
- plt.xlabel() and plt.ylabel() is used for labelling x and y axis

```python
import pandas as pd
import matplotlib.pyplot as plt

file= pd.ExcelFile("Example_Excel.xlsx")
print("Total sheets in Excel file is : {}".format(len(file.sheet_names)))
df = pd.read_excel("Example_Excel.xlsx", sheet_name= "Sheet1", index_col=None)
print(" Available columns in an Excel sheet1 \n {} \n".format(df.columns))
print("Columns count {}".format(len(df.columns)))
#df.shape
plt.bar(df["Country"], df["Gross Sales"])
plt.title("Countries VS Sales")
plt.xlabel("Countries")
plt.ylabel("Gross Sales")
plt.show()

#df["Gross Sales"].sum()
#Sorted_column_values = df.sort_values(['Gross Sales', 'Year'])
#Sorted_column_values
#Sorted_column_values['Gross Sales'].plot[Sorted_column_values['Year']]
#year=int(input("Enter the year:"))
#print(per_year(year))

#def per_year(year):
    #one_year_df = df[df["Year"]==year]
    #gross_sales_sum = one_year_df["Gross Sales"].sum()
   # print("Gross Sales of {} year is {}".format(year,gross_sales_sum)
    #return gross_sales_sum
   
```
