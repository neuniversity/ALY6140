# Read csv file and sort csv file
Read csv file and sort csv file based on two conditions: Intensity_IntegratedIntensityEdge_DAPI, Intensity_IntegratedIntensity_PI. The output is in ascendind order of Intensity_IntegratedIntensityEdge_DAPI, if the firgures are the same for Intensity_IntegratedIntensityEdge_DAPI, then sort based on Intensity_IntegratedIntensity_PI. Using this method, we can always get the smallest value of each column.   

```python
import pandas as pd
df = pd.read_csv('Nuclei.csv')
result = df.sort_values(by=['Intensity_IntegratedIntensityEdge_DAPI','Intensity_IntegratedIntensity_PI'])
print(result.head(4))
```

# Read the binary file.

```python
import pyabf
abf = pyabf.ABF("16o03002.abf")
print(abf)
```

# Access Sweep data and plot a sweep by using matplotlib.pyplot

```python
abf.setSweep(3)

import matplotlib.pyplot as plt
plt.figure(figsize=(80, 10)) #figsize represents length and width of the plot.
plt.plot(abf.sweepX, abf.sweepY, lw=6) #lw represents line width.
plt.show()
```
