### Example_Excel
Financial Sample file contains attributes like country, year, segmentation, product etc and measures like product sales, COGS(cost of good sold), units sold, profit etc. This data gives a overview about product sales, cost and profit along with other measures.This data was in the raw form and just by looking difficult to find meaningful insights out of it.

After reading the file using pandas library I found detailed information about how large is datasets in terms of columns and rows and using aggregation functions I was able to find meaningful insights which includes product units sold within each segment, profit and sales by country, units sold vs profit by product and Profit by country by year.


#### Below code is used to print first row within dataset.

```python
# This function is used to print first row within the dataset
import pandas as pd
df = pd.read_excel("Financial Sample.xlsx") # read_excel is used to read an Excel table into a pandas DataFrame
data = df.iloc[:1] # This function is used to print first row within the dataset
print(data)
df
```

#### Below code for displaying number of rows and columns in a dataset i.e. 700 rows and 16 columns.

#### Below I have used groupby function to show the number of units sold by segment and product wise. 5 segments are there in this dataset those are channel parthners, Enterprise, Goverment, Midmarket and small business.

```python
# Unit sold by Products within each segment
import pandas as pd
df=pd.read_excel("Financial Sample.xlsx")
units_sold=df.groupby(['Segment','Product'])['Units Sold'].sum()  # Group by function is used for aggregation
print(units_sold)
```

#### In below code I have calculated Gross sales and profit grouped by country to find out which country has highest product sales and profit. Based on data Unites States of America is making maximum sales however France is making maximum profit compared to other countries.

```python
# Gross Sales and Profit by Country
import pandas as pd
df=pd.read_excel("Financial Sample.xlsx")
print(df.groupby('Country')['Gross Sales','Profit'].sum())
```

#### The Below graph gives analysis of unit sold and profit by each product. Pareseo product has highest profit.

```python
# Units sold and Profit by products
import matplotlib.pyplot as plt
df=pd.read_excel("Financial Sample.xlsx")
print(df.groupby('Product')['Profit','Units Sold'].sum().plot.bar())
```

#### The below graph gives you an analysis of which country has earned maximum profit based on year 2013 and 2014. Here,Profitwise United states of america is one of the top most country in year 2013 and 2014.

```python
# To show profit by country in 2013 and 2014
import matplotlib.pyplot as plt
import pandas as pd

df.groupby(['Year','Country'])['Profit'].size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

### Example_Binary
.abf file is a Axon Binary format files created by data acquistion programs from Axon Instruments, Inc. Axon instruments are used by researchers for studying the electrical properties of biological cells which measures voltage and electrical changes and data produced during this is being recorded in the Axon Binary Format (ABF).

#### pyabf library is used for reading the data and it tells data contains how many channels (2 Channels), sweeps and length of recording (0.10 min).

```python
import pyabf
abf = pyabf.ABF("16o03002.abf") 
print(abf)
```

#### Below code for accessing sweep data

```python
# for accessing sweep data
abf.setSweep(0)
print(abf.sweepY) # display sweeps data
print(abf.sweepC) # display sweep commands (ADC)
print(abf.sweepX) # display sweep times (seconds)
```

```python
#### Below code for plotting sweep data using matplotlib.
```

```python
# for plotting sweep data
import matplotlib.pyplot as plt
import pyabf
abf = pyabf.ABF("16o03002.abf")
abf.setSweep(0)
plt.figure(figsize=(8,5))
plt.plot(abf.sweepX, abf.sweepY, lw=.5)
plt.show()
```

#### Based on file 2 channels data has been captured and below section tells about data by each channel where one represent membrane potential and other represent champ current by seconds.

```python
# This section displays graphs by channels and we have 2 channels within dataset.
import pyabf
abf = pyabf.ABF("16o03002.abf")

fig = plt.figure(figsize=(8, 5))

abf.setSweep(sweepNumber=0, channel=0)
ax1 = fig.add_subplot(211)
ax1.set_title(f"Channel {abf.sweepChannel+1}")
ax1.plot(abf.sweepX, abf.sweepY, lw=.5)
ax1.set_ylabel(abf.sweepLabelY)

abf.setSweep(sweepNumber=0, channel=1)
ax2 = fig.add_subplot(212)
ax2.set_title(f"Channel {abf.sweepChannel+1}")
ax2.plot(abf.sweepX, abf.sweepY, lw=.5)
ax2.set_xlabel(abf.sweepLabelX)
ax2.set_ylabel(abf.sweepLabelY)

fig.subplots_adjust(hspace=.4)  

plt.show()
```
