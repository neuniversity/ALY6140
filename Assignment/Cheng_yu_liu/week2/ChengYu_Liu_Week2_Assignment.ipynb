#### Cropping an Image:

```python
def main(): 
    try: 
        #Relative Path 
        img = Image.open("Nuclei_Msk.jpg") 
        width, height = img.size 
          
        area = (0, 0, width/2, height/2) 
        img = img.crop(area) 
          
        #Saved in the same relative location 
        img.save("Nuclei_Msk1.jpg")  
          
    except IOError: 
        pass
  
if __name__ == "__main__": 
    main() 
    
Image.open("Nuclei_Msk1.jpg")
```

#### This image is cropped. 

#### Rotating an Image:

```python
def main(): 
    try: 
        #Relative Path 
        img = Image.open("Nuclei_Msk.jpg")  
          
        #Angle given 
        img = img.rotate(180)  
          
         #Saved in the same relative location 
        img.save("Nuclei_Msk2.jpg") 
    except IOError: 
        pass
  
if __name__ == "__main__": 
    main()
Image.open("Nuclei_Msk2.jpg")
```

#### This image is rotated.

## Example_Binary:

#### The ABF format is also known as the "Adobe Glyph Bitmap Distribution Format."

```python
import pyabf
import matplotlib.pyplot as plt
```

```python
abf = pyabf.ABF("16o03002(1).abf")
print(abf.sweepY)
```

```python
abf.setSweep(4)
plt.figure(figsize=(8, 5))
plt.plot(abf.sweepX, abf.sweepY, lw=.5)
plt.show()
```
