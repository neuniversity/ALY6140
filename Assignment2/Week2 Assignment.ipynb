```python
import pyabf
df5=pyabf.ABF('16o03002.abf')
df5.setSweep(2)
plt.figure(figsize=(6,4))
plt.plot(df5.sweepX,df5.sweepY,lw=.5)
plt.show()
```

```python
import pyabf
def data_chart():
    df6=pyabf.ABF('16o03002.abf')
    plt.figure(figsize=(6,5))
    plt.plot(df6.sweepX,df6.sweepY,lw=.6)
    plt.ylabel(df6.sweepLabelY)
    plt.xlabel(df6.sweepLabelX)
    plt.title("Data Chart")
    return
data_chart()
```
