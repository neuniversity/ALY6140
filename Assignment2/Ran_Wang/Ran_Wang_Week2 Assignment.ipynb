```python
#IMPORT EXCEL
import pandas as pd
df = pd.read_excel('Financial Sample.xlsx',sheet_name = 'Sheet1')

#SORT EXCEL
df_index = df.set_index(['Product','Country','Discount Band']).sort_index()
df_index
```

```python
#PIVOT TABLE 
table = pd.pivot_table(df, values='Profit', index=['Product','Country','Discount Band'],
                        aggfunc={'Profit': [min, max, np.mean]})
table
```

```python
#Bar Plot 
table2 = table.xs(('Paseo','United States of America'), level=0)
table2.plot.bar()

plt.xlabel("Discount Band")
plt.ylabel("Profit")
plt.title('How Discount Band Impact on the Profit of Paseo in USA')
```

From the bar plot above, Paseo product will get maximum profit in low discount band compared with other discount bands in United States. If Paseo adopt a medium discount band, they will run the risk of losing money.

```python
#IMPORT ABF FILE
import matplotlib.pyplot as plt
import pyabf
abf = pyabf.ABF("16o03002.abf")
print(abf)

#PLOT THE DISTRIBUTION 
plt.figure(figsize=(8, 5))
for sweepNumber in range(abf.sweepCount)[0:6]:
    abf.setSweep(sweepNumber)
    plt.plot(abf.sweepX,abf.sweepY,alpha=.5,label=f"sweep {sweepNumber+1}")

plt.margins(0, .1)
plt.legend()
plt.ylabel(abf.sweepLabelY)
plt.xlabel(abf.sweepLabelX)
plt.title(abf.abfID)
plt.tight_layout()
plt.show()
```

```python
#EXTRACT EVERY SWEEP
abf = pyabf.ABF("16o03002.abf")
def run(a):
  i=0
  while i <= a : 
   abf.setSweep(i) # sweeps start at 0
   plt.plot(abf.sweepX, abf.sweepY, lw=.5)
   plt.show()
   i=i+1
run(5)
```

From the abf distribution plot, the membrane potential is stablized from sweep 1 to sweep 4 at the level around -56(mv) until it has reached a peak at 20(mv) from sweep 5 to sweep 6.
