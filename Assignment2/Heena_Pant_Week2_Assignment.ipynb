### Selective Reading of a CSV file.

The following block of code, displays `first row` from the top and `last row` from the bottom from the csv file.

> CSV file used: `Example_CSV.csv`

```python
def read_csv_file():
    df = pd.read_csv(
        "Example_CSV.csv"
    )

    df1 = df.iloc[:,0:3].head(1)#Selecting first 3 columns and 1 row from top
    df2 = df.iloc[:,0:3].tail(1)
    print(df1)
    print(df2)

read_csv_file()
```

### Selective reading of Excel File (Manipulation)

Selectively reading Excel file based on following conditions:
- `Country` is `Germany`
- `Profit` is `>10000`

Only displaying following columns:
- Segment
- Country
- Product
- Profit
- Units Sold

`Also, showing only top 3 rows.`

> Excel file used: `Example_Excel.xlsx`

```python
def read_excel_file():
    df = pd.read_excel(
        "Example_Excel.xlsx"
    )
    df1 = df.loc[(df["Country"] == "Germany") & (df["Profit"] > 10000)]

    print(df1[['Segment', 'Country', 'Product', 'Profit', 'Units Sold']].head(3))

read_excel_file()
```

### Line Chart Using Excel

The following code is to present a line chart, that shows total units sold for 3 countries in the excel file. The line grpah has following axis definitions:

- x axis: Country (Only Germany, Canada & USA included)
- y axis: Total number of Units Sold in those countries

> Excel file used: `Example_Excel.xlsx`

```python
import matplotlib.pyplot as plt

def read_excel_file():
    df = pd.read_excel(
        "Example_Excel.xlsx"
    )

    germany_df = df.loc[(df["Country"] == "Germany")]
    germany_df_units_sold = germany_df[['Units Sold']].sum()
    
    canada_df = df.loc[(df["Country"] == "Canada")]
    canada_df_units_sold = canada_df[['Units Sold']].sum()

    usa_df = df.loc[(df["Country"] == "United States of America")]
    usa_df_units_sold = usa_df[['Units Sold']].sum()
    
    mexico_df = df.loc[(df["Country"] == "Mexico")]
    mexico_units_sold = mexico_df[['Units Sold']].sum()
    
    france_df = df.loc[(df["Country"] == "France")]
    france_units_sold = france_df[['Units Sold']].sum()

    plt.plot(['Germany', 'Canada', 'USA', 'Mexico', 'France'], [germany_df_units_sold, canada_df_units_sold, usa_df_units_sold, mexico_units_sold, france_units_sold])
    plt.show()

read_excel_file()

```

### Manipulating Image File. 

Applying following operations on the given example image:

- Convert to `greyscale`
- Rotate `30` degrees

> Image used in this example is `Example_Image.jpg`

```python
 def read_image_file():
    coloredImage = Image.open("Example_Image.jpg")
    greyScaledImage = coloredImage.convert("L").rotate(30) #convert image to greyscale and rotate 30 degrees
    greyScaledImage.show()

read_image_file()
```

### Reading a binary file

Following code is to read a binary file, in our case an `abf` file

> Image used in this example is `Example_Binary.abf`

```python
import pyabf     #Import useful package
def read_binary_file():
    abf = pyabf.ABF("Example_Binary.abf")     #Read .abf file
    print(abf)

read_binary_file()
```

----

### Plotting Graph from a binary file

Following code is to plot a `X, Y graph` from a binary file using `matplotlib`, in our case an `abf` file

> Image used in this example is `Example_Binary.abf`

```python
def plot_binary_file():
    abf = pyabf.ABF("Example_Binary.abf")     #Read .abf file
    abf.setSweep(0)
    plt.figure(figsize=(8, 5))
    plt.plot(abf.sweepX, abf.sweepY, lw=.5)
    plt.show()

plot_binary_file()
```
