# Importing pandas library read csv file
Reading the head 10 rows of csv. We need to choose a plot to show data. Boxplot can show one column distribution well since those data ranged by column is disordered, making it clear to check is better.

````python
plt.figure(figsize=(10,5))
number1 = csv.loc[:,'Intensity_IntegratedIntensityEdge_DAPI'].head(10)
number2 = csv.loc[:,'Intensity_IntegratedIntensityEdge_PI'].head(10)
number3 = csv.loc[:,'Intensity_IntegratedIntensity_DAPI'].head(10)

plt.boxplot([number1,number2])
plt.gca().set_xticklabels(['Intensity_IntegratedIntensityEdge_DAPI','Intensity_IntegratedIntensityEdge_PI'])
````

# Showing distribution of data by boxplot
Using method pandas.DataFrame.loc to access a group of rows and columns by label, if I don't create the index, the default index usually is integer numbers012...by rows. The column index should be created as (:,'index name')
boxplot(x), x is what the data extracted is, other parameters in the method is default, like 'lables','widths',etc.

````python
exc = pd.read_excel('Financial Sample.xlsx')
# exc.head()
exc.groupby('Segment').mean()
# exc['Units Sold'].mean()
# df1['Units Sold']
# df2 = df1.groupby('Units Sold').mean()
# df2['Segment']
````

# Importing pandas library read excel file
The segment should be regard as a good index for showing the excel file, but there are some repeated characters, using method DataFrame.groupby( ) to divide the excel into several rows by unique segment, but if not using statistics method like mean( ), the error would be happened when creating new dataframe since other column attributes will conflict eache other by segment.

````python
Segment = ['Channel Partners','Enterprise','Government','Midmarket','Small Business']
df = exc.groupby('Segment').mean()
Unit_Sold = df['Units Sold']
plt.figure(figsize=(10,5))
plt.bar(Segment, Unit_Sold, 0.8)
plt.xlabel('Segment')
plt.ylabel('Units Sold')
plt.title('The mean of Units Sold for Segments')
````
