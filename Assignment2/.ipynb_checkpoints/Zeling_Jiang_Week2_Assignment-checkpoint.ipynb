# Image File

1. read image
2. make image gray
3. make image blurry and save as another file
4. check image's size
5. resize and rotate image

```python
from PIL import Image,ImageFilter
img = Image.open('Nuclei_Msk.jpg')
img
```

```python
gray = img.convert('L')
bw = gray.point(lambda x: 0 if x<128 else 255, '1')
bw.show()
```

```python
img_f = img.filter(ImageFilter.BLUR)
img_f.show()
img_f.save('blur.jpg')#save a new jpg
```

```python
img.size
```

```python
img_s = img.resize((200,100))
img_s
```

```python
img_r = img.rotate(45)
img_r
```

# Binary File

1. open and check information of abf file
2. plot single sweep
3. create all sweeps in one plot

```python
import pyabf
import matplotlib.pyplot as plt
abf = pyabf.ABF('16o03002.abf')
print(abf)#check information of abf file
```

```python
abf.setSweep(0)
print(abf.sweepY)
print(abf.sweepC)
print(abf.sweepX)
```

```python
abf.setSweep(0)#choose sweep from 0 to 5 and plot
plt.plot(abf.sweepX, abf.sweepY, lw=.3)#lw=line width
plt.show()
```

```python
plt.figure(figsize=(8,6))#plot size
for sweepNumber in range(abf.sweepCount)[0:6]:
    abf.setSweep(sweepNumber)
    plt.plot(abf.sweepX,abf.sweepY,alpha=.9,label=f'sweep{sweepNumber+1}')
    
plt.margins(0,.1)#range of Y
plt.legend()#can change name,location
plt.ylabel(abf.sweepLabelY)#defalt label or rename
plt.xlabel('time')
plt.title(abf.abfID)#ID-name of file
plt.show()
```
