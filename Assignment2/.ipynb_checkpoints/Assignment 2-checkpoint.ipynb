```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dealing with the excel file
datasetE =pd.read_excel('Financial Sample.xlsx') #Read the original data
Government =datasetE[datasetE.Segment=='Government']#Extract the governments' data
USGovernment = Government[Government.Country=='United States of America']# focus on the United States of America
USGovernment2014 = USGovernment[USGovernment.Year==2014]# focus on the year 2014
USGovernment2014Profit = USGovernment2014.Profit# Since I am interested in the profit, so I analyzed the profit seperately 
mean = round(USGovernment2014Profit.mean(),2)# Calculate the mean
std = round(USGovernment2014Profit.std(),2)# Calculate the standarddeviation
Max = round(max(USGovernment2014Profit),2)# Calculate the max
Min = round(min(USGovernment2014Profit),2)# Calculate the min
print('The average profit of US goverment is {0}$ with a standard deviation of {1}. Also, the maxinmum and minimum is {2}$ and {3}$.'.format(mean,std,Max,Min))
USGovernment2014Profit.hist(bins=10)# look at how it is distributed
```

```python
import pyabf
```

```python
#Dealing with the abf file:
abf = pyabf.ABF("16o03002 .abf")
print(abf)#To understand the basic charateristics of the file
def abfplot (a):# Since the steps are the same for 6 sweeps, setting a function will be a better solution to plot all of it
  i=0#first set the counter, also as the number of the sweep
  W=231#the spot for the plot
  plt.figure(figsize=(6, 4), dpi=150)#deciding the plot size
  while i <= a : # create the while loop, set the condition while the counter is smaller or equal to the sweep 
   plt.subplot(W)#assign the plot to a spot
   abf.setSweep(i) # sweeps start at 0
   plt.plot(abf.sweepX, abf.sweepY, lw=.5)
   plt.xticks([0,0.2,0.5,1])
   plt.title('sweep={}'.format(i+1))#set the title as the number of the sweep
   i=i+1
   W=W+1
  plt.subplots_adjust(wspace = 0.5,hspace = 0.5)#Create space between each plot
  plt.show()#The code to show the plot
# Since from print(abf) we learned that there are a total 6 sweeps in the file and python counts from 0, we set the number 5
abfplot(5)#From looking at the plot we can see they all have a peek around 0.2
```

```python
# After seeing all six sweep, I found that their y axises have different range. So I decide to put them all in one 
def abfplot2 (a):
  i=0#also  set the counter the number of the sweep
  plt.figure(figsize=(3, 3), dpi=200)#deciding the plot size
  while i <= a : # create the while loop, set the condition while the counter is smaller or equal to the sweep 
   abf.setSweep(i) # sweeps start at 0
   plt.plot(abf.sweepX, abf.sweepY, lw=.3,label='sweep={}'.format(i+1))#set the label as the sweep number
   i=i+1
  plt.xticks([0,0.2,0.5,1])
  plt.title("16o03002 .abf")#set the title
  plt.legend(loc='upper right')#set the legend
  plt.show()#The code to show the plot
abfplot2(5)#putting them all together, we can see that sweep 5 and 6 have the maximum value
```
