```python
import xlrd
loc = ("Financial Sample.xlsx")
wb = xlrd.open_workbook(loc)
sheet = wb.sheet_by_index(0) 
# print(sheet.row(0)) 
# print(sheet.row(1))
# print(sheet.nrows)
tmpmax =  sheet.cell_value(1,6)
for i in range(1, sheet.nrows):
    if sheet.cell_value(i,6) > tmpmax:
        tmpmax = sheet.cell_value(i, 6)
print("maximum sale price:", tmpmax)
```

I print first and second rows of the Excel workbook. I find the sixth cell of the first row is sale price. I want to find the maximum sale price of this sheet. So I write a forloop function. In order to find the range, I calcualte the numbers of row in this sheet and set it as the maximum index of the range. 

```python
import pyabf 
abf = pyabf.ABF("16o03002.abf")
print (abf)
abf.setSweep(3)
print(abf.sweepY)
print(abf.sweepC)
print(abf.sweepX)
```

```python
import matplotlib.pyplot as plt
import pyabf
abf=pyabf.ABF ("16o03002.abf")
abf.setSweep(4)
plt.plot (abf.sweepX,abf.sweepY,lw=0.5)
plt.show()
```
