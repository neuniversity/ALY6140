```python
import pyabf
import matplotlib.pyplot as plot
abf_plot = pyabf.ABF('16o03002.abf')
abf.setSweep(0)
plot.figure(figsize=(10, 5))
plot.plot(abf.sweepX, abf.sweepY, lw=.5)
plot.show()     #Plot the first sweep of .abf file
```

```python
import matplotlib.pyplot as plot2
import pyabf
abf = pyabf.ABF("16o03002.abf")

plot2.figure(figsize=(10, 5))
for sweepNumber in range(abf.sweepCount)[0:6]:
    abf.setSweep(sweepNumber)
    plt.plot(abf.sweepX,abf.sweepY,alpha=.5,label=f"sweep {sweepNumber+1}")
    
plot2.margins(0, .1)
plot2.legend()
plot2.ylabel(abf.sweepLabelY)
plot2.xlabel(abf.sweepLabelX)
plot2.title(abf.abfID)
plot2.tight_layout()
plot2.show()
```
