```python
import matplotlib.pyplot as plt
import pyabf
abf = pyabf.ABF("16o03002-binary.abf")

plt.figure(figsize=(5,1))
for sweepNumber in range(abf.sweepCount)[::1]:
    abf.setSweep(sweepNumber)
    plt.plot(abf.sweepX,abf.sweepY,alpha=.1, label=f"sweep {sweepNumber+1}")
    
plt.margins(0, .1)
plt.legend()
plt.ylabel (abf.sweepLabelY)
plt.ylabel (abf.sweepLabelX)
plt.title(abf.abfID)
plt.tight_layout()
plt.show()
```
