# #Extract the most values from the file

```python

import pandas as pd     #Import useful library
csvdf = pd.read_csv('Nuclei.csv')     #Read CSV file
csvdf_sorted = csvdf.sort_values('Intensity_IntegratedIntensityEdge_DAPI') #Sort CSV file by 'Intensity_IntegratedIntensityEdge_DAPI'
max = csvdf['Intensity_IntegratedIntensityEdge_DAPI'].max()     #Calculate values of interest
min = csvdf['Intensity_IntegratedIntensityEdge_DAPI'].min()
print('The maximum value under Intensity_IntegratedIntensityEdge_DAPI column is {0}'.format(max))
print('The minimum value under Intensity_IntegratedIntensityEdge_DAPI column is {0}'.format(min))
csvdf_sorted.head(n = 10)     #Display the first ten rows of CSV file after sorting, as well as values of interest
```

About the sample CSV file, CSVDF lists the first 10 lines of the original file. "Csvdf_sort" is generated according to "intensity ty_integratedintensity tyedge_dapi" to sort the original CSV file in ascending order. The manipulated CSV file also displays other information about the object, such as the corresponding object number and sequence number in the worksheet. The maximum, minimum and value of ty_integratedintensity edge_dapi were calculated and printed.

# #Access the scan data and draw the scan map

```python
import pyabf
abf = pyabf.ABF('16o03002.abf')
plt.plot(abf.sweepX, abf.sweepY)
plt.show()  #Plot the first sweep of .abf file
```

With respect to the abf sample file, "abf" displays basic information, including the number of channels, number of scans, and total length. The "plot" displays the first scan of the sample file. From the figure, we can see that the peak value is around 0.2 seconds, and other values of multiple scans are relatively similar.
