# Example_image
To read below image, I have considered three ways. 
1. resize: use image.resize() function;
2. change color type: image.convert(), where mode 'L' stands for greyscale;
3. rotating image: image.rotate() function;

```python
from PIL import Image
image = Image.open("Nuclei_Msk.jpg")
# image.show()
print(image.size) #get the image size

#use resize()method to change the size of this image
new_image = image.resize((300, 300))
new_image.save('Nuclei_Msk_image_300.jpg')
print(new_image.size) #output the new size of this image
new_image
```

```python
image = Image.open('Nuclei_Msk_image_300.jpg')
new_image = image.convert('L')
#new_image.save('Nuclei_Msk_image_300_greysacle_image.jpg')
new_image
```

```python
#Rotating images
image = Image.open("Nuclei_Msk.jpg")
image_rot_30 = image.rotate(30) #rotate image at 30 degrees
#image_rot_30.save('Nuclei_Msk_image_rot_30.jpg')
image_rot_30
```

In the above, I have rotated the image at 30 degrees. As we can see that the rotated image keep the original dimensions. Besides, the new image has cut its edges and padded to fit the original dimensions.

```python
image = Image.open("Nuclei_Msk.jpg")
image_rot_30_expand = image.rotate(18, expand = True)
#image_rot_30_expand.save('Nuclei_Msk_image_rot_30_expand.jpg')
image_rot_30_expand
```

To expand the dimensions of the rotated image to fit the entire the view, I have added expand=True in image.rorate()function. As we could see that the four edges of the image can be fully visible.

# Example_abf
read .abf file by importing package pyabf, which simplifies the process of reading data from Axon Binary Format(ABF). I was confused about abf.file at the beginning, and I have done some research and take a doc from https://github.com/swharden/pyABF/tree/master/docs/getting-started as a reference.
1. plot sweepNumber from 0 to 5 in one graph;
2. use subplot

```python
import pyabf
import matplotlib.pyplot as plt
abf = pyabf.ABF("16o03002.abf")
print(abf) #display basic information in .abf file
abf.setSweep(3) #6 sweeps means that the range of sweep is from 0 to 5
print(abf.sweepY) #sweep data 
print(abf.sweepC) #sweep command 
print(abf.sweepX) #sweep times (seconds)
```

```python
abf = pyabf.ABF("16o03002.abf")
plt.figure(figsize=(8, 5)) #set up the whole figure size

#plot sweepNumber from 0 to 5 in one graph
for sweepNumber in [0, 1, 2, 3, 4, 5]:
    abf.setSweep(sweepNumber)
    plt.plot(abf.sweepX, abf.sweepY, alpha=.8,
             label=f"sweep {sweepNumber}")
plt.margins(0, 0.1)
plt.legend() #added legend in interms of different sweepNumber
plt.ylabel(abf.sweepLabelY) #label y-axis
plt.xlabel(abf.sweepLabelX) #label x-axis
plt.title(abf.abfID) #added title
plt.show()
```

```python
import pyabf
import matplotlib.pyplot as plt
import numpy as np
abf = pyabf.ABF("16o03002.abf")
plt.figure(figsize=(15, 8))
#display each graph in terms of its corresponding sweepNumber, and plot in two rows, three columns
i=321
for sweepNumber in range(abf.sweepCount)[0:6]:
    abf.setSweep(sweepNumber)
    plt.subplot(i)
    plt.plot(abf.sweepX, abf.sweepY, lw=0.5)
    plt.xticks([0, 0.05, 0.1, 0.15, 0.2, 0.3, 0.4, 0.5, 0.8, 1.0])
    plt.title('Sweep{}'.format(sweepNumber+1))
    i=i+1
plt.subplots_adjust(wspace = 0.5, hspace = 0.5) #set up the space between each graph
```

From above graphs, we could see that the peak appears around 1.0 and 2.0.
