```python
import pyabf
df2 = pyabf.ABF("Desktop/Master/6140/Week 2/Assignment data/206.abf")
df2.setSweep(3)   #Sweeps start at 0
print(df2.sweepY) #Sweep data (ADC)
print(df2.sweepC) #Sweep Command (DAC)
print(df2.sweepX)   #Sweep times (Second)
```

```python
import matplotlib.pyplot as plt
import pyabf

df2 = pyabf.ABF("Desktop/Master/6140/Week 2/Assignment data/206.abf")
df2.setSweep(4)
plt.figure(figsize=(8,5))
plt.plot(df2.sweepX, df2.sweepY, lw=.5)
plt.ylabel(df2.sweepLabelY)
plt.xlabel(df2.sweepLabelX)
plt.title("Binary Chart")

plt.show()
```
