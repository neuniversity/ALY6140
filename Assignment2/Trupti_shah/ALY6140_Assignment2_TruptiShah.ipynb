### --------------- Image File ---------------

#### Reading a .jpg file:

```python
# Read image
image_jpg = ('/Users/Trupti/Desktop/Analytics/ALY6140/Week2Data/Nuclei_Msk.jpg')
image = Image.open(image_jpg)
#Display image
image.show()
```

#### Retrieving size of the given image - 

```python
with Image.open(image_jpg) as image: 
    width, height = image.size
print ('Width of the image is:', width)
print ('Height of the image is:', height)
```

#### Rotating the image -

```python
def rotate(): 
    try: 
        image = Image.open(image_jpg)
        # Angle given 
        image = image.rotate(180)  
        # Saved in the same relative location 
        image.save("rotated_image.jpg") 
    except IOError: 
        pass
  
if __name__ == "__main__": 
    rotate()
```

#### Resizing the image -

```python
def resize(): 
    try: 
        image = Image.open(image_jpg)
   
        img = image.resize((500, 400))  
        #Saved in the same relative location 
        img.save("resized_image.jpg")  
    except IOError: 
        pass
  
if __name__ == "__main__": 
    resize()
```

#### Cropping the image - 

```python
def crop(): 
    try: 
        image = Image.open(image_jpg) 
          
        area = (0, 0, width/2, height/2) 
        image = image.crop(area) 
          
        #Saved in the same relative location 
        image.save("cropped_image.jpg")  
          
    except IOError: 
        pass
  
if __name__ == "__main__": 
    crop()
```

#### Transposing the image - 

```python
def transpose(): 
    try: 
        image = Image.open(image_jpg) 
          
        #transposing image  
        transposed_img = image.transpose(Image.FLIP_LEFT_RIGHT) 
          
        #Save transposed image 
        transposed_img.save("transposed_image.jpg") 
    except IOError: 
        pass
  
if __name__ == "__main__": 
    transpose()
```

#### Pasting an image on another image - 

```python
def overlapping(): 
    try: 
        image1 = Image.open(image_jpg)  
          
        #Relative Path 
        #Image which we want to paste 
        image2_png = ('/Users/Trupti/Desktop/Analytics/ALY6140/Week2Data/image2.png')
        image2 = Image.open(image2_png)  
        image1.paste(image2, (50, 50)) 
          
        #Saved in the same relative location 
        image1.save("overlapped_image.jpg") 
        
        #Display image
        image1.show()
          
    except IOError: 
        pass
  
if __name__ == "__main__": 
    overlapping() 
```

### --------------- Binary File ---------------

#### Read an Axon Binary Format (.abf) file

```python
abf = pyabf.ABF('/Users/Trupti/Desktop/Analytics/ALY6140/Week2Data/16o03002.abf')
print(abf)
```

#### Access Sweep Data - 

```python
abf.setSweep(3) # sweeps start at 0
print(abf.sweepY) # sweep data (ADC)
print(abf.sweepC) # sweep command (DAC)
print(abf.sweepX) # sweep times (seconds)
```

#### Let's see some class variables - 

```python
DateTime = abf.abfDateTime
print(DateTime)
path = abf.abfFilePath
print(path)
dataRate = abf.dataRate
print(dataRate)
data = abf.dataSecPerPoint
print(data)
print(abf.sweepLabelX)
print(abf.sweepLabelY)
```

#### Plotting a Single Sweep - 

```python
abf.setSweep(3)
plt.plot(abf.sweepX, abf.sweepY, lw=.5)
plt.show()
```

#### Creating Plots with Matplotlib - 

```python
plt.figure(figsize=(8, 5))
for sweepNumber in range(abf.sweepCount)[::5]:
    abf.setSweep(sweepNumber)
    plt.plot(abf.sweepX,abf.sweepY,alpha=.5,label=f"sweep {sweepNumber+1}")

plt.margins(0, .1)
plt.legend()
plt.ylabel(abf.sweepLabelY)
plt.xlabel(abf.sweepLabelX)
plt.title(abf.abfID)
plt.tight_layout()
plt.show()
```

#### Plotting Multi-Channel ABFs - 

```python
fig = plt.figure(figsize=(8, 5))

abf.setSweep(sweepNumber=0, channel=0)
ax1 = fig.add_subplot(211)
ax1.set_title(f"Channel {abf.sweepChannel+1}")
ax1.plot(abf.sweepX, abf.sweepY, lw=.5)
ax1.set_ylabel(abf.sweepLabelY)

abf.setSweep(sweepNumber=0, channel=1)
ax2 = fig.add_subplot(212)
ax2.set_title(f"Channel {abf.sweepChannel+1}")
ax2.plot(abf.sweepX, abf.sweepY, lw=.5)
ax2.set_xlabel(abf.sweepLabelX)
ax2.set_ylabel(abf.sweepLabelY)

fig.subplots_adjust(hspace=.4)  # add more space between the subplots

plt.show()
```
