```python
import matplotlib.pyplot as plt # plt use to show image
import matplotlib.image as mpimg # mpimg use to read image
import numpy as np
```

```python
lena = mpimg.imread('Nuclei_Msk.jpg')
lena.shape
plt.imshow(lena) 
plt.axis('off') 
plt.show()
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

gray = rgb2gray(lena)
# Also can use plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.imshow(gray, cmap='Greys_r')
plt.axis('off')
plt.show()
```
