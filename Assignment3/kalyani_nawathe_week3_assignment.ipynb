```python
### GRADED
# (a) Determine the shape, number of dimensions and type of 
# elements in the numpy array arr
# (b) Determine the sum of elements in arr
# (c) Determine the mean of elements in arr
# (d) Create an array of the same shape as arr but filled with zeros
# (e) Create an array of the same shape as arr but filled with ones
# (f) Create an array of the same shape as arr but where all elements are squared values
# (g) Create an array result of shape 3*3 resulting from multiplication of arr with transpose(arr). Hint: use np.dot

arr = np.arange(15).reshape(3, 5)
print("Original array (arr):"'\n', arr)
```

```python
### GRADED
# The following piece of code creates a DataFrame 
# (a) Determine the shape of the dataframe
# (b) Print out all the columns of the dataframe
# (c) Print the 3rd element of the dataframe
# (d) Find the average of 'petal_width' where species is 'virginica'
# (e) Find the maximum of 'sepal_width' where species is 'setosa'
# (f): What is the average value of sepal_length
# (g): What is the maximum value of sepal_width
# (h): What is the minimum value of petal_width
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
df.head(10) #prints the first 10 rows
```

#### Utilize different functions and methods in pandas

```python
# (a) shape of the dataframe
print ("(a) Shape: ", df.shape)

# (b) columns of the dataframe
print ('\n'"(b) Columns: ",df.columns.values)

# (c) 3rd element of the dataframe
## this question is ambiguous, hence I consider 3rd element as 3rd column
## both .loc and .iloc can be used to take single element from the dataframe
third_elem_2 = df.loc[:, 'petal_length'] #by label
third_elem_1 = df.iloc[:, 2] #by location
print ('\n'"(c) Third element/column values (displaying only head): "'\n', third_elem_2.head())

## considering 3rd element as 3rd row of each column
third_row = df.iloc[2,:]
print('\n'"3rd element of row:  "'\n',third_row)

# (d) average of 'petal_width' where species is 'virginica'
## using groupby to calculate the mean for given column and row
print ('\n'"(d) Average of petal width for virginica species: %.2f" 
       % df.groupby('species').mean().loc['virginica','petal_width']) 

# (e) maximum of 'sepal_width' where species is 'setosa'
## procedure same as above question
print ('\n'"(e) Maximum of sepal width for setosa species: %.2f" 
       % df.groupby('species').max().loc['setosa','sepal_width']) 

# (f) average value of sepal_length
## using single element name from dataframe 
print ('\n'"(f) Average value of sepal length: %.2f" 
       % df['sepal_length'].mean()) 

# (g) maximum value of sepal_width
## this time using loc
print ('\n'"(g) Maximum value of sepal width: %.2f" 
       % df.loc[:, 'sepal_width'].max()) 

# (h) minimum value of petal_width
## this time using iloc
print ('\n'"(g) Minimum value of petal width: %.2f" 
       % df.iloc[:, 3].min())
```
