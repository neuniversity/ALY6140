```python
### GRADED
# (a) Determine the shape, number of dimensions and type of 
# elements in the numpy array arr
# (b) Determine the sum of elements in arr
# (c) Determine the mean of elements in arr
# (d) Create an array of the same shape as arr but filled with zeros
# (e) Create an array of the same shape as arr but filled with ones
# (f) Create an array of the same shape as arr but where all elements are squared values
# (g) Create an array result of shape 3*3 resulting from multiplication of arr with transpose(arr). Hint: use np.dot
arr = np.arange(15).reshape(3, 5)


# (a)Determine the shape, number of dimensions and type of 
# elements in the numpy array arr
print("(a)\narr")
print(arr)
print("="*10)
print("arr shape")
print(arr.shape)
print("="*10)
print("arr type")
print(type(arr))
print("="*10)
print("type of elements")
print(arr.dtype)

# (b) Determine the sum of elements in arr
print("\n(b)")
print("sum of elements in arr")
print(np.sum(arr))


# (c) Determine the mean of elements in arr
print("\n(c)")
print("mean of elements in arr")
print(np.mean(arr))

# (d) Create an array of the same shape as arr but filled with zeros
print("\n(d)")
print("same shape as arr but filled with zeros")
print(np.zeros_like(arr))

# (e) Create an array of the same shape as arr but filled with ones
print("\n(e)")
print("the same shape as arr but filled with ones")
print(np.ones_like(arr))

# (f) Create an array of the same shape as arr but where all elements are squared values
print("\n(f)")
print("he same shape as arr but where all elements are squared values")
print(np.full_like(arr,1.0,dtype=np.double))

# (g) Create an array result of shape 3*3 resulting from multiplication of arr with transpose(arr). Hint: use np.dot
print("\n(g)")
print("shape 3*3 resulting from multiplication of arr with transpose(arr)")
print(np.dot(arr,np.transpose(arr)))
```

```python
### GRADED
# The following piece of code creates a DataFrame 
# (a) Determine the shape of the dataframe
# (b) Print out all the columns of the dataframe
# (c) Print the 3rd element of the dataframe
# (d) Find the average of 'petal_width' where species is 'virginica'
# (e) Find the maximum of 'sepal_width' where species is 'setosa'
# (f): What is the average value of sepal_length
# (g): What is the maximum value of sepal_width
# (h): What is the minimum value of petal_width
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
print(df.head(10)) #prints the first 10 rows

# (a) Determine the shape of the dataframe
print("\n(a)")
print("the shape of the dataframe")
print(df.shape)

# (b) Print out all the columns of the dataframe
print("\n(b)")
print("all the columns of the dataframe")
print(df.columns)

# (c) Print the 3rd element of the dataframe
print("\n(c)")
print("the 3rd element of the dataframe")
print(df.iloc[2,])

# (d) Find the average of 'petal_width' where species is 'virginica'
print("\n(d)")
print("average of 'petal_width' where species is 'virginica'")

a=df[df['species']== 'virginica']
print(a['petal_width'].mean())


# (e) Find the maximum of 'sepal_width' where species is 'setosa'
print("\n(e)")
print("the maximum of 'sepal_width' where species is 'setosa'")
b=df[df['species']== 'setosa']
print(b['sepal_width'].max())

# (f): What is the average value of sepal_length
print("\n(f)")
print("average value of sepal_length")
print(df['sepal_length'].mean())

# (g): What is the maximum value of sepal_width
print("\n(g)")
print("the maximum value of sepal_width")
print(df['sepal_width'].max())

# (h): What is the minimum value of petal_width
print("\n(h)")
print("the minimum value of petal_width")
print(df['petal_width'].min())
```
