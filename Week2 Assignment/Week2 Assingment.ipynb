```python
import matplotlib.pyplot as plt
import pyabf
abf = pyabf.ABF("16o03002.abf") #Read the file
plt.plot(abf.sweepX, abf.sweepY, lw=.5) #Make the plot that sweepX and Y. 
plt.figure(figsize=(8, 5)) #Set the range of figure size
for sweepNumber in range(abf.sweepCount)[0:6]: #Use the loop(for) to select the ranges of 6 sweeps
    abf.setSweep(sweepNumber)
    plt.plot(abf.sweepX,abf.sweepY,alpha=.5,label=f"sweep {sweepNumber+1}") 
plt.legend(loc='best',fontsize= 'medium')  #Set the legend and put the best location in the plot with med size 
plt.title(abf.abfID) #Create the title of the plot graph
plt.xlabel(abf.sweepLabelX) #Create the xlabel (TIme(Second)
plt.ylabel(abf.sweepLabelY) #Create the ylabel (Membrane Potential)
plt.show() # Display the plot
```

```python
import pandas as pd
import numpy as np
import xlrd
data = pd.read_excel("Financial Sample.xlsx") #Import the Excel file
Enterprise = data[data.Segment=='Enterprise'] #Bring a part of data (Enterprise)
CaEnterprise = Enterprise[Enterprise.Country=='Canada'] #Extract the Canadian Enterprise part 
CaEnterprise13 = CaEnterprise[CaEnterprise.Year==2013] # Input a specific condition (Year)
CaEnterprise13Profit = CaEnterprise13.Profit #Input theprofit part from the data
std = round(CaEnterprise13Profit.std(),2) #Print the standard deviation for figuring out the average profit's variety
mean = round(CaEnterprise13Profit.mean(),2) #Calculate the medium number
Min = round(min(CaEnterprise13Profit),2) #Figure the minimum number out
Max = round(max(CaEnterprise13Profit),2) #Figure the maximum number out
print (Max) #Print the maximum value
print (Min) #Print the minimum
print (mean) #Print the medium figure
print (std) #Calculate the std value
print('The average profit of 2013 Canada Enterprise is {0}$ with a standard deviation of {1}. Also, the maxinmum and minimum is {2}$ and {3}$.'.format(mean,std,Max,Min)) #Comprehensively show the Canadian Enterprises' average profit in 2013 from the mean,std,max,min
CaEnterprise13Profit.hist(bins=10) #Display the histogram of Canadian enterprise avg profit in 2013
```
